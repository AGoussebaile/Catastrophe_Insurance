```python
import matplotlib.pyplot as plt
import pandas as pd

#data = pd.read_csv('1nVar.csv') #comparative static on capital cost c
#data = pd.read_csv('2etaVar.csv') #comparative static on risk aversion eta
#data = pd.read_csv('3corVar.csv') #comparative static on risk correlation rho
data = pd.read_csv('4proVar.csv') #comparative static on probability p and loss L

ES_VaR_ref = data['14'].iloc[18]/data['13'].iloc[18]
ES_VaR_k_eta__ref = data['15'].iloc[18]/data['13'].iloc[18]

data['1'] = data['1']*100
data['7'] = data['7']*100
data['13'] = data['13']*100
data['19'] = data['19']*100
data['25'] = data['25']*100

data['2'] = data['2']/ES_VaR_ref*100
data['8'] = data['8']/ES_VaR_ref*100
data['14'] = data['14']/ES_VaR_ref*100
data['20'] = data['20']/ES_VaR_ref*100
data['26'] = data['26']/ES_VaR_ref*100

data['3'] = data['3']/ES_VaR_k_eta__ref*100
data['9'] = data['9']/ES_VaR_k_eta__ref*100
data['15'] = data['15']/ES_VaR_k_eta__ref*100
data['21'] = data['21']/ES_VaR_k_eta__ref*100
data['27'] = data['27']/ES_VaR_k_eta__ref*100

data['4'] = data['4']*100
data['10'] = data['10']*100
data['16'] = data['16']*100
data['22'] = data['22']*100
data['28'] = data['28']*100

data['5'] = data['5']*100
data['11'] = data['11']*100
data['17'] = data['17']*100
data['23'] = data['23']*100
data['29'] = data['29']*100

data['6'] = data['6']*100
data['12'] = data['12']*100
data['18'] = data['18']*100
data['24'] = data['24']*100
data['30'] = data['30']*100
data
```
